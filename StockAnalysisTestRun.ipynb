```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
```

```python
start = '2021-01-01'
end = '2021-08-08'
```

```python
google = data.DataReader("GOOG", 'yahoo', start, end)
google.head()
```

```python
google['Open'].plot(label = 'GOOG Open Price', figsize=(15, 7))
google['Close'].plot(label = 'GOOG Close Price')
plt.legend()
plt.title('Google Stock Prices')
plt.ylabel('Stock Price')
```

```python
google['Volume'].plot(figsize=(17,5))
```
